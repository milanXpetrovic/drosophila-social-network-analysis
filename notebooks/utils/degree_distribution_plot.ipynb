```python
import networkx as nx

import os
import sys
import numpy as np
import pandas as pd
from src import settings
import matplotlib.pyplot as plt
from src.utils import fileio

def degree_frequency(degrees):
    unique_degrees, counts = np.unique(degrees, return_counts=True)
    return unique_degrees, counts

def degree_probability(degrees):
    unique_degrees, counts = np.unique(degrees, return_counts=True)
    probabilities = counts / sum(counts)  # Normalize counts to get probabilities
    return unique_degrees, probabilities


config = fileio.get_config(settings.CONFIG_NAME)
data_path = "/srv/milky/drosophila-datasets/drosophila-isolation/data/processed/2_2_create_total_graph"

for TREATMENT in config["TREATMENTS"]:
    graphs = fileio.load_files_from_folder(os.path.join(data_path, TREATMENT), '.gml')

    degree_count_all = []
    degree_total_interaction_times_all = []
    for g_name, g_path in graphs.items():
        g = nx.read_gml(g_path)
        degree_count = [deg for _, deg in g.degree(weight='count')]
        degree_total_interaction_times = [int(deg) for _, deg in g.degree(weight='total_interaction_times')]    
        degree_count_all.append(degree_count)
        degree_total_interaction_times_all.append(degree_total_interaction_times)

    degree_count_values, degree_count_freq = degree_probability(degree_count_all)
    degree_tit_values, degree_tit_freq = degree_probability(degree_total_interaction_times_all)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(degree_count_values, degree_count_freq, color='blue', alpha=0.7, edgecolor='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Count {TREATMENT}')
    plt.xlabel('Weighted Degree')
    plt.ylabel('Frequency')
    plt.xlim(1, 1500)
    plt.ylim(0, 1)

    plt.subplot(1, 2, 2)
    plt.scatter(degree_tit_values, degree_tit_freq, color='red', alpha=0.7, edgecolor='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Total Interaction Times {TREATMENT}')
    plt.xlabel('Weighted Degree')
    plt.ylabel('Frequency')
    plt.xlim(0, 1500)
    plt.ylim(0, 1)

    plt.tight_layout()
    plt.show()

    # sys.exit()
```

```python
path = "/srv/milky/drosophila-datasets/drosophila-isolation/data/processed/1_0_find_interactions/CS_10D"

edgellists = fileio.load_files_from_folder(path)

# key, value = next(iter(edgellists.items()))
# edgellists = {key: value}

thresholds = [1*60*24, 3*60*24, 5*60*24, 10*60*24, 20*60*24]

for threshold in thresholds:
    degree_count_all = []
    degree_total_interaction_times_all = []

    for name, path in edgellists.items():
        df = pd.read_csv(path,index_col=0)
        # df = df[['node_1', 'node_2', 'start_of_interaction']]
        df = df.reset_index(drop=True)

        df_interactions = df.sort_values("start_of_interaction")
        df_interactions = df[df['start_of_interaction'] < threshold]
        G = nx.DiGraph()

        for _, row in df_interactions.iterrows():
            node_1, node_2 = row["node_1"], row["node_2"]
            duration = row["duration"] / config["FPS"]

            if G.has_edge(node_1, node_2):
                G[node_1][node_2]["count"] += 1
                G[node_1][node_2]["total_interaction_times"] += duration

            else:
                G.add_edge(node_1, node_2, count=1, total_interaction_times=duration)
        
        degree_count = [deg for _, deg in G.degree(weight='count')]
        degree_total_interaction_times = [int(deg) for _, deg in G.degree(weight='total_interaction_times')]    
        degree_count_all.append(degree_count)
        degree_total_interaction_times_all.append(degree_total_interaction_times)

    degree_count_all_new = []
    for l in degree_count_all:
        if len(l) < 12:
            for _ in range(12 - len(l)):
                l.append(0)
        
        degree_count_all_new.append(l)

    degree_total_interaction_times_all_new = []
    for l in degree_total_interaction_times_all:
        if len(l) < 12:
            for _ in range(12 - len(l)):
                l.append(0)
        
        degree_total_interaction_times_all_new.append(l)


    degree_count_values, degree_count_freq = degree_probability(degree_count_all_new)
    degree_tit_values, degree_tit_freq = degree_probability(degree_total_interaction_times_all)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(degree_count_values, degree_count_freq, color='blue', alpha=0.7, edgecolor='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Count {TREATMENT} {threshold/(60*24)}')
    plt.xlabel('Weighted Degree')
    plt.ylabel('Frequency')
    max_val = max(max(inner_list) for inner_list in degree_count_all_new if inner_list)
    plt.xlim(-10, max_val + 0.2*max_val)
    plt.ylim(0, 1)

    plt.subplot(1, 2, 2)
    plt.scatter(degree_tit_values, degree_tit_freq, color='red', alpha=0.7, edgecolor='black')
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Total Interaction Times {TREATMENT} {threshold}')
    plt.xlabel('Weighted Degree')
    plt.ylabel('Frequency')
    max_val = max(max(inner_list) for inner_list in degree_total_interaction_times_all if inner_list)
    plt.xlim(-10, max_val+0.2*max_val)
    plt.ylim(0, 1)

    plt.tight_layout()
    plt.show()
```
