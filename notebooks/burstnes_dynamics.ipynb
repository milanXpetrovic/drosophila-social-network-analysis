```python
import pandas as pd
from src.utils import fileio
import matplotlib.pyplot as plt

path = "/home/milky/drosophila-SNA/data/processed/CsCh/1_0_find_edges"

files = fileio.load_files_from_folder(path)

for file_name, file_path in files.items():
    df = pd.read_csv(file_path, index_col=0)
    values = list(range(0, 43200))

    event_locations  = df["start_of_interaction"].to_list()
    event_duration = df["duration"].to_list()
    plt.figure(figsize=(14, 1))
    plt.hlines(1, 0, 43200)  # Horizontal line at y=1 to represent the events

    for l, d in zip(event_locations, event_duration):
        plt.axvline(x=l, linestyle='-', alpha=0.01, linewidth=d)

    plt.xlabel('Time')
    plt.ylabel('Height')
    plt.title(file_name)
    plt.xlim(0, 43200)
    plt.ylim(0, 1)

    plt.show()
```
