```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

def find_saddle_point(data):
    peaks, _ = find_peaks(data)
    if len(peaks) < 2:
        return None

    saddle_index = np.argmin(data[peaks[0]:peaks[1]]) + peaks[0]

    return saddle_index

data = np.array([0, 0, 0, 0, 0, 1, 1, 1, 1, 2, 3, 4, 5, 4, 3, 2, 1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]) 
saddle_index = find_saddle_point(data)

plt.plot(data)
peak_index = np.argmax(data)
plt.axvline(x=saddle_index, color='red', linestyle='--')
plt.show()
```
